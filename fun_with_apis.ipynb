### International Space Station Location and Time API

```python
import requests
```

```python
response = requests.get('http://api.open-notify.org/iss-now.json')
print(response.status_code)
```

```python
import json
```

```python
response = requests.get('http://api.open-notify.org/iss-now.json')
print(response.json())
```

```python
json_response = response.json()
response_dict = json.dumps(json_response
                           , sort_keys = True
                           , indent = 4
                          )
print(response_dict)
```

```python
json_response['timestamp']
```

```python
key = 'e1cd236585b1260e1977fdb8cb1a98bf' # rest, get your own whenever you run this
lat = '37.7749' # San Francisco latitde 37 North
long = '-122.4194' # San Francisco longitude 122 West
exclude = 'minutely,hourly,daily,alerts,flags'
darksky_url = 'https://api.darksky.net/forecast/'+ key + '/' + lat + ',' + long + '?exclude=' + exclude

darksky_response = requests.get(darksky_url)
darksky_dict = json.dumps(darksky_response.json()
                          , sort_keys = True
                          , indent = 4
                         )
print(darksky_dict)
```

### Frankfurt Stock Exchange API

```python
import modin.pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
FSE_response = requests.get('https://www.quandl.com/api/v3/datasets/FSE/BDT_X')
FSE_dict = json.dumps(FSE_response.json()
                      , sort_keys = True
                      , indent = 4
                     )
print(FSE_dict)
```

```python
FSE_response_json = FSE_response.json()
FSE_df = pd.DataFrame(FSE_response_json['dataset']['data']
                        , columns = FSE_response_json['dataset']['column_names']
                       )
FSE_df.head()
```
